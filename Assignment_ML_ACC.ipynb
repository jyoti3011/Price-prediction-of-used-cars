```python
# Replace '?' with NaN
df.replace('?', np.nan, inplace=True)
```

```python
# Convert relevant columns to numeric
numeric_cols = ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Drop rows with missing target values
df.dropna(subset=['price'], inplace=True)
```

```python
# Impute missing values with mean for numerical columns
for col in ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm']:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python

# Impute missing values with mode for categorical columns
df['num-of-doors'].fillna(df['num-of-doors'].mode()[0], inplace=True)
```

```python
df.dtypes
```

```python
 #Convert Columns to Integer (int64)

int_columns = ['ID', 'symboling', 'curb-weight', 'engine-size', 'city-mpg', 'highway-mpg']
df[int_columns] = df[int_columns].astype('int64')
```

```python
#Convert Columns to Float (float64)

float_columns = ['normalized-losses', 'wheel-base', 'length', 'width', 'height',
                 'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm']
df[float_columns] = df[float_columns].astype('float64')
```

```python
 #Convert Columns to Object (Categorical)

object_columns = ['num-of-doors', 'engine-type', 'num-of-cylinders', 'fuel-system']
df[object_columns] = df[object_columns].astype('object')
```

```python
# Identify columns with prefix 'make_', 'fuel-type_', etc.
bool_columns = [col for col in df.columns if col.startswith(('make_', 'fuel-type_', 'aspiration_', 'body-style_', 'drive-wheels_', 'engine-location_'))]
df[bool_columns] = df[bool_columns].astype('bool')
```

```python
corr_matrix = df.corr(numeric_only=True)
```

```python
# Get absolute correlations with 'price' and sort
top_features = corr_matrix['price'].abs().sort_values(ascending=False).head(10).index
```

```python
# Compute the correlation matrix for top features
top_corr = df[top_features].corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(top_corr, dtype=bool))
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(top_corr, mask=mask, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Top Correlated Features with Price')
plt.show()
```

```python
make_columns = [col for col in df.columns if col.startswith('make_')]
```

```python


average_prices = {}
for make in make_columns:
    make_name = make.replace('make_', '').replace('-', ' ').title()
    average_price = df.loc[df[make] == True, 'price'].mean()
    average_prices[make_name] = average_price
```

```python
avg_price_df = pd.DataFrame(list(average_prices.items()), columns=['Make', 'Average Price'])
avg_price_df = avg_price_df.sort_values(by='Average Price', ascending=False).dropna()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Make', y='Average Price', data=avg_price_df)
plt.xticks(rotation=45)
plt.title('Average Price by Car Make')
plt.ylabel('Average Price')
plt.xlabel('Car Make')
plt.tight_layout()
plt.show()
```

```python
#Steps to Calculate Drive-Wheel Proportions

# Assuming 'df' is your DataFrame
total_cars = len(df)
drive_counts = {
    'FWD': df['drive-wheels_fwd'].sum(),
    'RWD': df['drive-wheels_rwd'].sum(),
    '4WD': df['drive-wheels_4wd'].sum()
}

drive_proportions = {k: v / total_cars for k, v in drive_counts.items()}

# Convert to DataFrame for better visualization
drive_df = pd.DataFrame({
    'Drive Type': drive_counts.keys(),
    'Count': drive_counts.values(),
    'Proportion': drive_proportions.values()
})

print(drive_df)
```

```python
# Steps to Count Cars with Four Doors

# Count the number of cars with four doors
four_door_count = df[df['num-of-doors'] == 'four'].shape[0]
print(f"Number of cars with four doors: {four_door_count}")
```

```python
# Steps to Calculate the Proportion of Zeros in symboling


# Count of zero entries in 'symboling'
zero_count = (df['symboling'] == 0).sum()

# Total number of entries
total_count = df['symboling'].count()

# Proportion of zeros
zero_proportion = zero_count / total_count

print(f"Proportion of zeros in 'symboling': {zero_proportion:.2%}")
```
